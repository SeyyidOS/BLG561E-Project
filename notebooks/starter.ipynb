```python
import pandas as pd
import numpy as np
from itertools import combinations
from sklearn.preprocessing import OneHotEncoder
from scipy.spatial.distance import cosine, euclidean

data = df.copy()

players = data[['id', 'age', 'player_name', 'position', 'nationality', 'height', 'weight'] + list(data.columns[8:])]


players['age'] = players['age'].astype(float)

categorical_features = ['position', 'nationality']
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_categorical = encoder.fit_transform(players[categorical_features])
players['encoded_cat'] = list(encoded_categorical)

team_columns = list(data.columns[8:]) 
def extract_team_years(row):
    team_years = []
    for col, value in row.items():
        if pd.notnull(value) and value != '0':  
            team_years.append((value, col)) 
    return team_years

players['team_years'] = players[team_columns].apply(extract_team_years, axis=1)

numerical_features = ['age', 'height', 'weight']
players['numerical'] = players[numerical_features].values.tolist()

def enhanced_similarity(player1, player2, weights, sigma=1.0):
    w1, w2, w3 = weights

    try:
        if len(player1['encoded_cat']) != len(player2['encoded_cat']):
            cat_sim = 0
        else:
            cat_sim = 1 - cosine(player1['encoded_cat'], player2['encoded_cat'])
    except ValueError:
        cat_sim = 0

    num1 = np.array(player1['numerical'])
    num2 = np.array(player2['numerical'])
    valid = ~np.isnan(num1) & ~np.isnan(num2)
    if valid.any():
        num_sim = np.exp(-euclidean(num1[valid], num2[valid])**2 / (2 * sigma**2))
    else:
        num_sim = 0

    team_years_1 = set(player1['team_years'])
    team_years_2 = set(player2['team_years'])
    overlap = team_years_1.intersection(team_years_2)
    total = team_years_1.union(team_years_2)
    played_together_sim = len(overlap) / len(total) if total else 0

    return w1 * cat_sim + w2 * num_sim + w3 * played_together_sim

weights = (0.4, 0.4, 0.2)  

pairs = list(combinations(players.iterrows(), 2))
enhanced_pairs = []
for player1, player2 in pairs:
    idx1, p1 = player1
    idx2, p2 = player2
    score = enhanced_similarity(p1, p2, weights)
    enhanced_pairs.append((p1['player_name'], p2['player_name'], score))

# Convert pairs to a DataFrame
enhanced_pairs_df = pd.DataFrame(enhanced_pairs, columns=['Player 1', 'Player 2', 'Similarity Score'])

# Display top results
top_pairs = enhanced_pairs_df.sort_values(by="Similarity Score", ascending=False).head(10)
print(top_pairs)
```
